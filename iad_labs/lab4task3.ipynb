```python
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
```

```python
#Задание 1
X = np.arange(6).reshape(3, 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = linear_model.LinearRegression()
model.fit(X_poly, np.array([3, 2, 1]))
```

```python
X_test = np.linspace(0, 5, 100).reshape(50, 2)
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)
```

```python
plt.scatter(X[:, 0], [3, 2, 1], color='red', label='Исходные данные')
plt.plot(X_test[:, 0], y_pred, label='Полиномиальная регрессия')
plt.legend()
plt.show()
```

```python
#Задание 2
x = np.arange(5)
y = 3 - 2*x + x**2 - x**3
```

```python
pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=3)),
    ('linear', linear_model.LinearRegression())
])
```

```python
pipeline.fit(x[:, np.newaxis], y)

coef = pipeline.named_steps['linear'].coef_
print("Коэффициенты полинома:", coef)
```
